# TESTING

```python
#Define policy and rewards
SIZE = 13
pi = np.zeros((SIZE**2,SIZE**2))

for i in range(1, SIZE**2-1):
    u = i-SIZE
    d = i+SIZE
    l = i-1
    r = i+1
    if i%SIZE==SIZE-1:
        r = i
    if i+SIZE>SIZE**2-1:
        d = i
    if i%SIZE==0:
        l = i
    if i-SIZE<0:
        u = i
        
    pi[i,u]=pi[i,d]=pi[i,l]=pi[i,r]=0.25
    if u==r:
        pi[i,i] = 0.5
    if d==l:
        pi[i,i] = 0.5
        
rewards = np.full((SIZE**2, SIZE**2), -1.)
```

```python
delta_min = 1e-5
gamma = 0.9
```

```python
V = np.zeros(SIZE**2)
delta = 1.0
while delta > delta_min:
    delta=0
    for i in range(1,SIZE**2-1):
        old_V = deepcopy(V)
        V[i] = np.sum(pi[i]*(rewards[i]+gamma*old_V))
        delta = max(delta,np.sum(np.abs(old_V-V)))
V=V.reshape(SIZE,SIZE)
```

```python
def plot_V(V):
    #plt.figure(figsize=(3, 3))
    plt.pcolormesh(np.zeros_like(V), edgecolors='k', cmap="Greys", linewidth=1)
    ax = plt.gca()
    plt.xticks([])
    plt.yticks([])
    for i in range(V.shape[0]):
        for j in range(V.shape[1]):
            plt.text(j+0.5, V.shape[0]-0.5-i, f"{V[i,j]:.2f}", ha='center', va='center', fontsize=5)
```

```python
plot_V(V)
```

```python
def plot_grid(V):
    #plt.figure(figsize=(3, 3))
    layout = np.zeros_like(V)
    layout[0,V.shape[1]-1]=layout[V.shape[0]-1,0]=1
    plt.pcolormesh(layout, edgecolors='k', cmap="Greys", linewidth=2)
    ax = plt.gca()
    plt.xticks([])
    plt.yticks([])

    action_names = [r"$\uparrow$",r"$\downarrow$", r"$\leftarrow$", r"$\rightarrow$"]
    tot_grid = np.full((V.shape[0]+2, V.shape[1]+2), np.min(V)-10)
    tot_grid[1:-1,1:-1]=V
    q = np.zeros((V.shape[0], V.shape[1], 4))
    for i in range(1,V.shape[0]+1):
        for j in range(1,V.shape[0]+1):
            q[i-1,j-1,0] = tot_grid[i-1,j]
            q[i-1,j-1,1] = tot_grid[i+1,j]
            q[i-1,j-1,2] = tot_grid[i,j-1]
            q[i-1,j-1,3] = tot_grid[i,j+1]
    greedy_actions = np.argmax(q, axis=2)

    for i in range(V.shape[0]):
        for j in range(V.shape[1]):
            if (i==0 and j==0) or (i==V.shape[0]-1 and j==V.shape[1]-1):
                continue
            action_name = action_names[greedy_actions[i,j]]
            plt.text(j+0.5, V.shape[0]-0.5-i, action_name, ha='center', va='center')
```

```python
plot_grid(V)
```

# TESTING MAZE

```python
#Define policy and rewards
SIZE = 13
pi = np.zeros((SIZE**2,SIZE**2))

for i in range(0, SIZE**2-1):
    u = i-SIZE
    d = i+SIZE
    l = i-1
    r = i+1
    if i%SIZE==SIZE-1:
        r = l
    if i+SIZE>SIZE**2-1:
        d = u
    if i%SIZE==0:
        l = r
    if i-SIZE<0:
        u = d
        
    pi[i,u]=pi[i,d]=pi[i,l]=pi[i,r]=0.25
    pi[i]/=np.sum(pi[i])
        
rewards = a.flatten()
for i in range(0, SIZE**2-1):
    if np.abs(np.sum(pi[4]*rewards)+100)<1e-5:
        continue
    pi[i][pi[i]*rewards < -20]=0
    if np.sum(pi[i]) > 0:
        pi[i]/=np.sum(pi[i])
```

```python
delta_min = 1e-5
gamma = 1.0
```

```python
V = np.zeros(SIZE**2)
delta = 1.0
while delta > delta_min:
    delta=0
    for i in range(0,SIZE**2-1):
        old_V = deepcopy(V)
        V[i] = np.sum(pi[i]*(rewards+gamma*old_V))
        delta = max(delta,np.sum(np.abs(old_V-V)))
V=V.reshape(SIZE,SIZE)
```

```python
plot_V(V)
```
