# Plotting functions

```python
def plot_V(V):
    size = int(np.sqrt(len(V)))
    V = V.reshape(size, size)
    plt.pcolormesh(np.zeros_like(V), edgecolors='k', cmap="Greys", linewidth=1)
    ax = plt.gca()
    plt.xticks([])
    plt.yticks([])
    for i in range(V.shape[0]):
        for j in range(V.shape[1]):
            plt.text(j+0.5, V.shape[0]-0.5-i, f"{V[i,j]:.2f}", ha='center', va='center', fontsize=5)

def plot_policy(pi):
    grid_size = int(np.sqrt(len(pi[0])))
    layout = np.zeros((grid_size, grid_size))
    layout[0,grid_size-1]=layout[grid_size-1,0]=1
    plt.pcolormesh(layout, edgecolors='k', cmap="Greys", linewidth=2)
    ax = plt.gca()
    plt.xticks([])
    plt.yticks([])

    action_names = [r"$\uparrow$",r"$\downarrow$", r"$\leftarrow$", r"$\rightarrow$"]
    greedy_actions = np.zeros((grid_size**2), dtype=int)
    for state in range(1, (grid_size**2)-1):
        action = np.where(pi[state]==1.0)[0].item()
        if action == state-grid_size:
            greedy_actions[state] = 0
        elif action == state+grid_size:
            greedy_actions[state] = 1
        elif action == state-1:
            greedy_actions[state] = 2
        elif action == state+1:
            greedy_actions[state] = 3
        else:
            sys.exit("Policy is not valid!")
    greedy_actions = greedy_actions.reshape(grid_size, grid_size)

    for i in range(grid_size):
        for j in range(grid_size):
            if (i==0 and j==0) or (i==grid_size-1 and j==grid_size-1):
                continue
            action_name = action_names[greedy_actions[i,j]]
            plt.text(j+0.5, grid_size-0.5-i, action_name, ha='center', va='center')
            
            
def plot_grid(V):
    size = int(np.sqrt(len(V)))
    V = V.reshape(size, size)
    layout = np.zeros_like(V)
    layout[0,V.shape[1]-1]=layout[V.shape[0]-1,0]=1
    plt.pcolormesh(layout, edgecolors='k', cmap="Greys", linewidth=2)
    ax = plt.gca()
    plt.xticks([])
    plt.yticks([])

    action_names = [r"$\uparrow$",r"$\downarrow$", r"$\leftarrow$", r"$\rightarrow$"]
    tot_grid = np.full((V.shape[0]+2, V.shape[1]+2), np.min(V)-10)
    tot_grid[1:-1,1:-1]=V
    q = np.zeros((V.shape[0], V.shape[1], 4))
    for i in range(1,V.shape[0]+1):
        for j in range(1,V.shape[0]+1):
            q[i-1,j-1,0] = tot_grid[i-1,j]
            q[i-1,j-1,1] = tot_grid[i+1,j]
            q[i-1,j-1,2] = tot_grid[i,j-1]
            q[i-1,j-1,3] = tot_grid[i,j+1]
    greedy_actions = np.argmax(q, axis=2)

    for i in range(V.shape[0]):
        for j in range(V.shape[1]):
            if (i==0 and j==0) or (i==V.shape[0]-1 and j==V.shape[1]-1):
                continue
            action_name = action_names[greedy_actions[i,j]]
            plt.text(j+0.5, V.shape[0]-0.5-i, action_name, ha='center', va='center')

def plot_maze_solution(greedy_actions, old_maze, start_i=0, start_j=0):
    grid_size = old_maze.shape[0]
    old_maze = old_maze.flatten()
    maze = deepcopy(old_maze)
    
    state = start_j+grid_size*start_i
    while state != len(old_maze)-1:
        maze[state] = -50
        state = greedy_actions[state]
    maze[state] = -50
    
    #Plot maze together with solution
    f, mazes = plt.subplots(1,2)
    mazes[0].imshow(old_maze.reshape(grid_size, grid_size), cmap='Greys_r')
    mazes[1].imshow(maze.reshape(grid_size, grid_size), cmap = 'Greys_r')
    
def generate_maze(grid_size):
    np.random.seed(2)
    maze=np.zeros(shape=(grid_size, grid_size))
    start_point=np.array([0,0])
    maze_generator=Maze(maze, start_point)
    big_maze = maze_generator.maze
    #big_maze[grid_size-2,grid_size-1]=big_maze[grid_size-1,grid_size-1]=2
    big_maze[big_maze==2.] = -1
    big_maze[big_maze==0.] = -100
    plt.imshow(big_maze, cmap = 'Greys_r')
    return big_maze

def map_V_to_pi(V, rewards, pi_allowed):
    best_policy = list()
    V = np.zeros(grid_size**2)
    theta = 1e-24
    gamma = 0.95
    delta = 10
    while delta > theta:
        v = deepcopy(V)
        for i in range(0,len(V)-1):
            V[i] = np.sum(pi[i]*(rewards[i]+gamma*v))
        delta = np.sum(np.abs(v-V))
    V=V.reshape(grid_size, grid_size)
```

# Policy evaluation simple gridworld

```python
GAMMA = 1
SIZE = 4
DELTA = 1e-5
```

```python
def get_policy(size):
    """
    Get policy for gridworld, Sutton-Barto chapter 4
    
    Input : size - the side length of the square world
    Output: pi   - the policy sized size**2 x size**2
                   which for each state lists prob of moving 
                   to other state. 
    """
    pi = np.zeros((size**2, size**2))
    for state in range(1, (size**2)-1):
        left_state = state-1
        right_state = state+1
        up_state = state-size
        down_state = state+size
        
        if state==size-1:
            change_inds = np.asarray([left_state, state, down_state])
            pi[state][change_inds] = 0.25, 0.5, 0.25
            continue
        elif state==(size**2)-size:
            change_inds = np.asarray([right_state, state, up_state])
            pi[state][change_inds] = 0.25, 0.5, 0.25   
            continue
            
        if state%size==size-1:
            right_state = state
        elif state%size==0:
            left_state = state
        elif up_state<0:
            up_state = state
        elif down_state>size**2-1:
            down_state = state
            
        change_inds = np.asarray([up_state, down_state, left_state, right_state])
        pi[state][change_inds] = 0.25
        
    return pi

def get_rewards(size):
    #Size x size array
    rewards = np.full((size**2), -1.)
    return rewards
```

```python
rewards = get_rewards(SIZE)
pi = get_policy(SIZE)
diff = 1.0

V = np.zeros((SIZE**2))
while diff>DELTA:
    for i in range(1, len(V)-1):
        v = deepcopy(V)
        V[i] = np.sum(pi[i]*(rewards+GAMMA*v))
    
    diff = np.linalg.norm(v-V)
```

# Policy iteration

```python
def get_possible_actions_for_state(s, grid_size, grid_move=False):
    """
    Maps from current state to possible actions.
    For gridworld where actions are up/down/left/right.
    """
    up = s-grid_size
    down = s+grid_size
    left = s-1
    right = s+1
    if up<0:
        up = s
    if down>=grid_size**2:
        down = s
    if s%grid_size==0:
        left = s
    if s%grid_size==grid_size-1:
        right = s
        
    actions = np.asarray([up, down, left, right])
    if grid_move == False:
        return actions
    else:
        return actions[actions!=s]
    
def get_pi_for_state(state, V, rewards, GAMMA, choose_one=False):
    pi_state = np.zeros(len(V))
    grid_size = int(np.sqrt(len(V)))
    possible_actions = get_possible_actions_for_state(state, grid_size)
    vs_for_as = V[possible_actions]
    rs_for_as = rewards[possible_actions]
    sum_values = rs_for_as + GAMMA*vs_for_as
    if choose_one==False:
        max_as = np.where(sum_values==np.max(sum_values))[0]
        best_as = possible_actions[max_as]
        for best_a in best_as:
            pi_state[best_a] += 1.0/len(best_as)
    else:
        best_a = possible_actions[np.argmax(sum_values)]
        pi_state[best_a] = 1.0
    return pi_state
```

```python
def get_policy_from_V(V, rewards, GAMMA, choose_one=False):
    pi = np.zeros((len(V), len(V)))
    for state in range(1, len(V)-1):
        pi_for_state = get_pi_for_state(state, V, rewards, GAMMA, choose_one)
        pi[state] = pi_for_state
    return pi

def policy_evaluation(V, pi, rewards, GAMMA, DELTA):
    diff = DELTA+1.0
    while diff>DELTA:
        for i in range(1, len(V)-1):
            v = deepcopy(V)
            V[i] = np.sum(pi[i]*(rewards+GAMMA*v))
        diff = np.linalg.norm(v-V)
    return V

def policy_iteration(rewards, GAMMA, DELTA):
    V = np.zeros(len(rewards))
    pi = get_policy_from_V(V, rewards, GAMMA, choose_one=False)
    old_pi = np.ones_like(pi)
    while (old_pi==pi).all()==False:
        old_pi = deepcopy(pi)
        V = policy_evaluation(V, pi, rewards, GAMMA, DELTA)
        pi = get_policy_from_V(V, rewards, GAMMA, choose_one=True)
    return V, pi
```

```python
SIZE = 13
rewards = np.full(SIZE**2, -1.)
rewards[53:57] = -1000.
V, pi = policy_iteration(rewards, GAMMA, DELTA)
```
