```python
import numpy as np
import matplotlib.pyplot as plt
from copy import deepcopy
from maze_solver import dynamic_prog, utils, q_learning
import time
```

# Demonstrate maze solving with dynamic programming

```python
GAMMA = 0.95 #discount rate
DELTA = 1e-5 #stopping condition for policy evaluation.
maze = np.loadtxt('mazes/maze_35.csv')
plt.imshow(maze, cmap = 'Greys_r')
```

```python
# Use policy iteration to solve the maze
start = time.time()
V, pi = dynamic_prog.policy_iteration(maze, GAMMA, DELTA)
end = time.time()
print(f'Time elapsed: {end-start}')
```

```python
# Plot the maze solution, starting from point (5,0)
utils.plot_maze_solution(pi, maze, 5, 0)
```

# Solve maze with Q learning

```python
ALPHA = 0.8 # alpha parameter in Q learning
EPSILON = 0.1 # amount of exploration in e-greedy learning
GAMMA = 1.0 # discount factor
DELTA = 1e-5 # max difference between old and new Q as stopping condition
```

```python
# Use Q learning to solve the maze
start = time.time()
Q = q_learning.solve(maze, GAMMA, ALPHA, EPSILON, DELTA)
end = time.time()
print(f'Time elapsed: {end-start}')
```

```python
# Plot using Q from starting point 0, 30
utils.plot_solution_from_q(Q, maze, start_x=0, start_y=30)
```
