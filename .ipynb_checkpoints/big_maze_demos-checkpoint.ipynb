```python
import numpy as np
import matplotlib.pyplot as plt
from copy import deepcopy
from maze import Maze
import time
```

```python
def plot_V(V):
    #plt.figure(figsize=(3, 3))
    plt.pcolormesh(np.zeros_like(V), edgecolors='k', cmap="Greys", linewidth=1)
    ax = plt.gca()
    plt.xticks([])
    plt.yticks([])
    for i in range(V.shape[0]):
        for j in range(V.shape[1]):
            plt.text(j+0.5, V.shape[0]-0.5-i, f"{V[i,j]:.2f}", ha='center', va='center', fontsize=5)

def plot_grid(V):
    #plt.figure(figsize=(3, 3))
    layout = np.zeros_like(V)
    layout[0,V.shape[1]-1]=layout[V.shape[0]-1,0]=1
    plt.pcolormesh(layout, edgecolors='k', cmap="Greys", linewidth=2)
    ax = plt.gca()
    plt.xticks([])
    plt.yticks([])

    action_names = [r"$\uparrow$",r"$\downarrow$", r"$\leftarrow$", r"$\rightarrow$"]
    tot_grid = np.full((V.shape[0]+2, V.shape[1]+2), np.min(V)-10)
    tot_grid[1:-1,1:-1]=V
    q = np.zeros((V.shape[0], V.shape[1], 4))
    for i in range(1,V.shape[0]+1):
        for j in range(1,V.shape[0]+1):
            q[i-1,j-1,0] = tot_grid[i-1,j]
            q[i-1,j-1,1] = tot_grid[i+1,j]
            q[i-1,j-1,2] = tot_grid[i,j-1]
            q[i-1,j-1,3] = tot_grid[i,j+1]
    greedy_actions = np.argmax(q, axis=2)
    print(greedy_actions)

    for i in range(V.shape[0]):
        for j in range(V.shape[1]):
            if (i==0 and j==0) or (i==V.shape[0]-1 and j==V.shape[1]-1):
                continue
            action_name = action_names[greedy_actions[i,j]]
            plt.text(j+0.5, V.shape[0]-0.5-i, action_name, ha='center', va='center')

def plot_maze_solution(greedy_actions, old_maze, start_i=0, start_j=0):
    grid_size = old_maze.shape[0]
    old_maze = old_maze.flatten()
    maze = deepcopy(old_maze)
    
    state = start_j+grid_size*start_i
    while state != len(old_maze)-1:
        maze[state] = -50
        state = greedy_actions[state]
    maze[state] = -50
    
    #Plot maze together with solution
    f, mazes = plt.subplots(1,2)
    mazes[0].imshow(old_maze.reshape(grid_size, grid_size), cmap='Greys_r')
    mazes[1].imshow(maze.reshape(grid_size, grid_size), cmap = 'Greys_r')
    
def generate_maze(grid_size):
    np.random.seed(2)
    maze=np.zeros(shape=(grid_size, grid_size))
    start_point=np.array([0,0])
    maze_generator=Maze(maze, start_point)
    big_maze = maze_generator.maze
    #big_maze[grid_size-2,grid_size-1]=big_maze[grid_size-1,grid_size-1]=2
    big_maze[big_maze==2.] = -1
    big_maze[big_maze==0.] = -100
    plt.imshow(big_maze, cmap = 'Greys_r')
    return big_maze

def map_V_to_pi(V, rewards, pi_allowed):
    best_policy = list()
    V = np.zeros(grid_size**2)
    theta = 1e-24
    gamma = 0.95
    delta = 10
    while delta > theta:
        v = deepcopy(V)
        for i in range(0,len(V)-1):
            V[i] = np.sum(pi[i]*(rewards[i]+gamma*v))
        delta = np.sum(np.abs(v-V))
    V=V.reshape(grid_size, grid_size)
```

```python
def possible_actions_state(s, grid_size, grid_move = False):
    """
    Maps from current state to possible actions.
    For gridworld where actions are up/down/left/right.
    """
    up = s-grid_size
    down = s+grid_size
    left = s-1
    right = s+1
    if up<0:
        up = s
    if down>=grid_size**2:
        down = s
    if s%grid_size==0:
        left = s
    if s%grid_size==grid_size-1:
        right = s
        
    actions = np.asarray([up, down, left, right])
    if grid_move == False:
        return actions
    else:
        return actions[actions!=s]
    
def possible_actions(grid_size, grid_move = False):
    poss_as = list()
    for i in range(grid_size**2 - 1):
        poss_as.append(possible_actions_state(i, grid_size, grid_move))
    return poss_as

def get_best_actions(V, rewards, state, poss_a, gamma, grid_size, grid_move=False):
    """
    Input: V - state-value fn
           rewards
           state - current state
           gamma - discount rate
           grid_size - size of square gridworld
    Output: all best actions for a single state given V, as
            numpy array.
    """
    pi_state = np.zeros(len(V))
    next_state_values = rewards[poss_a] + gamma*V[poss_a]
    return poss_a[np.flatnonzero(next_state_values==np.max(next_state_values)).flatten()]

def get_pi_from_V(V, rewards, poss_as, gamma, grid_size, choose_one=False):
    """
    Input: V - state-value fn
           rewards
           gamma - discount rate
           grid_size - side size of square gridworld
           choose_one - decides to choose one action or output
                        probabilities for each best action.
    Output: policy based on V values.
    """
    pi = list()
    for i in range(len(V)-1):
        pi_state = np.zeros(len(V))
        best_states = get_best_actions(V, rewards, i, poss_as[i], gamma, grid_size, grid_move=True)
        if choose_one==False:
            for state in best_states:
                pi_state[state] += 1./len(best_states)
        else:
            best_state = best_states[0]#np.random.choice(best_states)
            pi_state[best_state] = 1.
        pi.append(pi_state)
    return np.asarray(pi)
```

```python
def policy_evaluation(V, pi, rewards, gamma, tol):
    """
    Inputs: pi - policy
            rewards - grid_size x grid_size matrix
            gamma - discount rate
            tol - tolerance (how close successive Vs should be)
    Output: V - state value fn for pi
    """
    delta = 10
    while delta > tol:
        v = deepcopy(V)
        for state in range(0,len(V)-1):
            V[state] = np.sum(pi[state]*(rewards+gamma*v))
        delta = np.sum(np.abs(v-V))
    return V

def policy_improvement(V, rewards, poss_as, gamma, grid_size):
     return get_pi_from_V(V, rewards, poss_as, gamma, grid_size, choose_one=True)
    
def get_actions_from_pi(pi):
    actions = list()
    for i in range(pi.shape[1]-1):
        actions.append(np.random.choice(np.arange(pi.shape[1]), p=pi[i]))
    return np.asarray(actions)

def value_iteration(V, rewards, gamma, grid_size, tol):
    delta = 10.
    while delta > tol:
        v = deepcopy(V)
        for state in range(0,len(V)-1):
            poss_a = possible_actions_state(state, grid_size, grid_move=True)
            V[state] = np.max(rewards[poss_a]+gamma*v[poss_a])
        delta = np.sum(np.abs(v-V))
    return V
```

# Policy iteration

```python
gamma = 0.95
theta = 1e-12
V = np.zeros(grid_size**2)     
r = np.full(grid_size**2, -1.)
poss_as = possible_actions(grid_size)
pi = get_pi_from_V(V, r, poss_as, gamma, grid_size)
new_actions = get_actions_from_pi(pi)
poss_as = possible_actions(grid_size, grid_move = True)
```

```python
#Iterative policy evaluation
start = time.time()
old_actions = np.zeros_like(new_actions)
while np.array_equal(new_actions, old_actions) == False:
    
    #Policy evaluation
    V = policy_evaluation(V, pi, big_maze.flatten(), gamma, tol=theta)
    old_actions = deepcopy(new_actions)

    #Policy improvement
    pi = policy_improvement(V, big_maze.flatten(), poss_as, gamma, grid_size)
    new_actions = get_actions_from_pi(pi)
V=V.reshape(grid_size, grid_size)
end = time.time()
print(f"Time: {end-start}")
```

```python
greedy_actions = get_actions_from_pi(pi)
plot_maze_solution(greedy_actions, big_maze, 0, 0)
```

```python
theta = 1e-12
gamma = 0.95
V = np.zeros(grid_size**2)
```
